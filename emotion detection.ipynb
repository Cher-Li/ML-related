```python
from transformers import pipeline

# Load pre-trained emotion detection model
# Model: j-hartmann/emotion-english-distilroberta-base
# Supports emotions like joy, sadness, anger, fear, surprise, disgust
emotion_classifier = pipeline(
    "text-classification",
    model="j-hartmann/emotion-english-distilroberta-base",
    return_all_scores=True
)

def detect_emotion(text):
    results = emotion_classifier(text)[0]  # Get results for first (only) text
    # Sort emotions by score, highest first
    sorted_results = sorted(results, key=lambda x: x['score'], reverse=True)
    top_emotion = sorted_results[0]
    return top_emotion['label'], top_emotion['score'], sorted_results

# ==== DEMO ====
print("Emotion Detector (type 'quit' to exit)")
while True:
    user_input = input("Enter a sentence: ")
    if user_input.lower() == "quit":
        break
    emotion, score, details = detect_emotion(user_input)
    print(f"Predicted Emotion: {emotion} (Confidence: {score:.2f})")
    print("Full Scores:", details)
```
